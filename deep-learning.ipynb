```python
#IMPORTACIONES

import numpy

import keras

import tensorflow as tf
```

```python
#LIBRERIA DE KERAS PARA PROCESAR IMAGENES

from keras.preprocessing import image
```

```python
'''IMPORTAMOS:
INCEPTIONV3, MODELO INCORPORADO POR KERAS PARA LA CLASIFICACION DE IMAGENES
DECODE_PREDICTIONS, PERMITE MAPEAR LAS CLASES'''


from keras.applications.inception_v3 import InceptionV3, decode_predictions
```

```python
#

iv3 = InceptionV3()
```

```python
#CONSTRUIMOS TENSOR DE ENTRADA 

x = image.img_to_array(image.load_img("/kaggle/input/iais22-birds/submission_test/submission_test/1033986.jpg", target_size=(299, 299)))

#load_img carga la imagen
#traget_size especifica ancho y alto para cargar la imagen (299, 299) ya que asÃ­ se especifica en la API y no son validos otros valores
#img_to_array convierte la imagen en un array de tres dimensiones --> altura, anchura y rgb
```

```python
#ajustamos los valores rgb al intervalo [-1,1] (estaban en el intervalo [0,255])

x /= 255
x -= 0.5
x *= 2
```

```python
#REAJUSTAMOS LAS 4 DIMENSIONES QUE DEBE TENER EL INCEPTIONV3, DANDOLE COMO PRIMER PARAMETRO 1 YA QUE LA ENTRADA CONTIENE 1 IMAGEN

x = x.reshape([1, x.shape[0], x.shape[1], x.shape[2]]) 
```

```python
#NUESTRO MODELO ANALIZA LA IMAGEN

y = iv3.predict(x)
```

```python
#DECODE_PREDICTIONS DEVUELVE LAS 5 PRIMERAS POSICIONES DE UNA MATRIZ ORIGINARIA DE 1x1000


print(decode_predictions(y))

#OBTENEMOS TRES ELEMENTOS POR TUPLA, EL ULTIMO NOS INDICA LA PROBABILIDAD Y EL ELEMENTO EN EL INDICE 1 NOS DEVUELVE EL NOMBRE POR EL CUAL SE HA CLASIFICADO LA IMAGEN
```
